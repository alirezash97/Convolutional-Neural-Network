```
import cv2 as cv 
import numpy as np
import matplotlib.pyplot as plt

image = cv.imread('/content/1_LEa6_2OjW-zUy8crecnY7A.jpeg')
plt.imshow(image)
plt.show
```

```
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
plt.imshow(gray)
plt.show
print(gray.shape)
```

```
x = gray.shape[0]
y = gray.shape[1]

kernel = np.array([[23, -10, 11],
                   [15, 90, 0],
                   [-87, -189, 96]])

u = kernel.shape[0]
v = kernel.shape[1]

stride = 2

strided = np.zeros(( int((x - u  + 1)/stride), int((y - v + 1)/stride)))


for i in range(strided.shape[0]):
  for j in range(strided.shape[1]):

    strided[i][j] = np.sum(gray[ i*stride : i*stride + u, j*stride : j*stride + v] * kernel)  

```

```
print(strided.shape)
plt.imshow(strided)
plt.show
```
