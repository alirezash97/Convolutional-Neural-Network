```
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np


image = cv.imread('/content/1_LEa6_2OjW-zUy8crecnY7A.jpeg') 
plt.imshow(image)
plt.show
print(image.shape)
```

```
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
plt.imshow(gray)
plt.show
print(gray.shape)
```

```
 

x = gray.shape[0]
y = gray.shape[1]

vertical_kernel = np.array([[1, 0, -1],
                            [1, 0, -1],
                            [1, 0, -1]])

horizontal_kernel = np.array([[1, 1, 1],
                              [0, 0, 0],
                              [-1, -1, -1]])

vertical_u = vertical_kernel.shape[0]
vertical_v = vertical_kernel.shape[1]

horizontal_u = horizontal_kernel.shape[0]
horizontal_v = horizontal_kernel.shape[1]

vertical_edge = np.zeros((x - vertical_u  + 2 ,y - vertical_v + 2))
horizontal_edge = np.zeros((x - horizontal_u  + 2 ,y - horizontal_v + 2))


#horizontal edge detection
for i in range(x - horizontal_u):
  for j in range(y - horizontal_v):

    horizontal_edge[i][j] = np.sum(gray[i:i + horizontal_u, j:j + horizontal_v] * horizontal_kernel)  


# vertical edge  detection
for i in range(x - vertical_u):
  for j in range(y - vertical_v):

    vertical_edge[i][j] = np.sum(gray[i:i + vertical_u, j:j + vertical_v] * vertical_kernel)



```

```
plt.imshow(vertical_edge)
plt.show
```
