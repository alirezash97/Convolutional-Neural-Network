```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg


image = mpimg.imread('/content/072319_ee_cat-allergy_feat.jpg')
plt.imshow(image)
plt.show
```

```
kernel_3d = np.array([[[1, 1, 1],   #detection on red channel
                      [0, 0, 0],
                      [-1, -1, -1]],
                     [[0, 0, 0],       #detection on green channel
                      [0, 0, 0],
                      [0, 0, 0]],
                     [[0, 0, 0],        #detection on blue channel 
                      [0, 0, 0],
                      [0, 0, 0]]])

x = image.shape[0]
y = image.shape[1]
u = kernel_3d.shape[1]
v = kernel_3d.shape[2]


output = np.zeros((x - u + 1, y - v + 1))

for i in range(output.shape[0]):
  for j in range(output.shape[1]):
    
    output[i][j] = (np.sum(image[i:i+u, j:j+v, :1] * kernel_3d[:1, :, :]) 
                 + np.sum(image[i:i+u, j:j+v, 1:2] * kernel_3d[1:2, :, :])
                 + np.sum(image[i:i+u, j:j+v, 2:3] * kernel_3d[2:3, :, :])) 
```

```
plt.imshow(output)
plt.show
```
