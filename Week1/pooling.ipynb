```
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

image = mpimg.imread('/content/072319_ee_cat-allergy_feat.jpg')
plt.imshow(image)
plt.show
```

```

def pooling(input_image, f, s, type):

  x = input_image.shape[0]
  y = input_image.shape[1]
  z = input_image.shape[2]

  output = np.zeros(( int((x - f) / s) + 1 , int((y - f) / s) + 1, z ))

  if type == 'max':
    
    for i in range(output.shape[2]):
      for j in range(output.shape[1]):
        for k in range(output.shape[0]):
          x = np.array(input_image[k*s:k*s+f, j*s:j*s+f, i])
          if x.shape[0] != 0:
            output[k][j][i] = np.max(x)


  elif type == 'average':
    
    
    for i in range(output.shape[2]):
      for j in range(output.shape[1]):
        for k in range(output.shape[0]):
          x = np.array(input_image[k*s:k*s+f, j*s:j*s+f, i])
          if x.shape[0] != 0:
            output[k][j][i] = np.mean(x)


  return output
```

```
a = pooling(image, 6, 3, 'max')
plt.imshow((a).astype(np.uint8))
plt.show
```

```
a = pooling(image, 6, 3, 'average')
plt.imshow((a).astype(np.uint8))
plt.show
```
