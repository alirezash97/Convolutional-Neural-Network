```
# Padding 
import cv2 as cv 
import matplotlib.pyplot as plt
import numpy as np


image = cv.imread('/content/image.jpeg')
plt.imshow(image)
plt.show
```

```
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
plt.imshow(gray)
plt.show
```

```
x = gray.shape[0]
y = gray.shape[1]

# 200px padding
padding_size = 200
half = int(padding_size / 2)
new_image = np.zeros((x + padding_size, y + padding_size))
new_image[half:x+half, half:y+half] = gray
plt.imshow(new_image)
plt.show 
```
